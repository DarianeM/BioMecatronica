```python
# Importar las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import sklearn
from sklearn import linear_model
```

```python
# Cargar el archivo de Excel
file_path = '/content/altura-peso.xlsx'
data = pd.read_excel(file_path)
print(data)
```

```python
# Mostrar las primeras filas del DataFrame
print(data.head())
```

```python
# Mostrar la información general del DataFrame
print(data.info())
```

```python
# Separar las variables independientes (peso) y dependientes (altura)
X = data[['Peso']]
y = data[['Altura']]

#grafiacs corresponditees de dispersión
plt.scatter(X,y)
plt.xlabel('Peso')
plt.ylabel('Altura')
plt.show()
```

```python
#Separo los datos de "train " en entrenamento y prueba para probar el algoritmo
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split (X,y, test_size=0.3)
# 80/20 test_size=0.2
```

```python
#defino el algoritmo a utilizar
from sklearn.linear_model import LinearRegression
```

```python

lr =linear_model.LinearRegression()
lr.fit(X_train, y_train)
```

```python
#PREDICCIÓN
Y_pred=lr.predict(X_test)
print(Y_pred)
```

```python
# Agregar la columna de predicción al DataFrame original
data['Prediccion'] = lr.predict(X)
print(data)
```

```python
# Evaluar el modelo
mse = sklearn.metrics.mean_squared_error(y_test, Y_pred)
r2 = sklearn.metrics.r2_score(y_test, Y_pred)
print(f'Error cuadrático medio (MSE): {mse}')
print(f'Coeficiente de determinación (R^2): {r2}')
```

```python
#grafiacs corresponditees de dispersión
plt.scatter(X_test, y_test, color='blue', label='Datos de prueba')
plt.plot(X_test, Y_pred, color='red', linewidth=3, label='Predicción')
plt.title('regresion lineal simple')
plt.xlabel('Peso')
plt.ylabel('Altura')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print('LA ECUACION ES  IGUAL')
print(f'Y = {lr.coef_[0][0]} X + {lr.intercept_[0]}')
```

```python
# Gráfica con los datos originales y las predicciones
plt.scatter(data['Peso'], data['Altura'], color='green', label='Datos originales')
plt.plot(data['Peso'], data['Prediccion'], color='red', linewidth=2, label='Predicción')
plt.title('Altura vs Peso - Datos originales y Predicción')
plt.xlabel('Peso')
plt.ylabel('Altura')
plt.legend()
plt.grid(True)
plt.show()
```
