## 2. Funciones y exploracion visual
Prepara una o más funciones que permitan realizar una exploración visual de los
datos más relevantes utilizando las librerías python de tu elección.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Definición de funciones para la exploración visual de los datos

def plot_histogram(data, column, title, bins=10, color='blue'):
    """
    Función para graficar histogramas de una columna especificada.

    Args:
    data (DataFrame): El DataFrame de pandas.
    column (str): La columna del DataFrame a graficar.
    title (str): Título del gráfico.
    bins (int): Número de divisiones del histograma.
    color (str): Color del histograma.
    """
    plt.figure(figsize=(10, 6))
    sns.histplot(data[column], bins=bins, color=color, kde=True)
    plt.title(title)
    plt.xlabel(column)
    plt.ylabel('Frecuencia')
    plt.grid(True)
    plt.show()
# Ejemplo de uso de una función para visualizar la edad
plot_histogram(data, 'age', 'Distribución de Edad')
```

```python
def plot_bar(data, column, title):
    """
    Función para graficar gráficos de barras de la distribución de valores de una columna especificada.

    Args:
    data (DataFrame): El DataFrame de pandas.
    column (str): La columna del DataFrame a graficar.
    title (str): Título del gráfico.
    """
    plt.figure(figsize=(10, 6))
    sns.countplot(x=column, data=data)
    plt.title(title)
    plt.ylabel('Frecuencia')
    plt.show()

# Ejemplo de uso de una función para visualizar la edad
plot_bar(data, 'age', 'Distribución de Edad')
```

```python
def plot_correlation_matrix(data, title):
    """
    Función para graficar la matriz de correlación de todas las columnas numéricas en el DataFrame.

    Args:
    data (DataFrame): El DataFrame de pandas.
    title (str): Título del gráfico.
    """
    plt.figure(figsize=(12, 10))
    sns.heatmap(data.corr(), annot=True, fmt=".2f", cmap='coolwarm')
    plt.title(title)
    plt.show()

# Ejemplo de uso de una función para visualizar la edad
plot_correlation_matrix(data, 'age')
```

## **3. Imputación y normalización de datos**
Si lo estimas necesario, ejecuta métodos de imputación y normalización de
datos explicando explicando brevemente tus decisiones.   

### Imputación

Mediana para variables continuas

```python
from sklearn.impute import SimpleImputer

# Crear un imputador que utilice la mediana para llenar los valores faltantes
imputer_median = SimpleImputer(strategy='median')

# Lista de columnas continuas que necesitan imputación de la mediana
continuous_columns = ['cigsPerDay', 'totChol', 'BMI', 'heartRate', 'glucose']

# Aplicar la imputación de la mediana a las columnas especificadas
data[continuous_columns] = imputer_median.fit_transform(data[continuous_columns])

# Verificar que no quedan valores faltantes en las columnas imputadas
print(data[continuous_columns].isnull().sum())
```

Moda para variables categóricas

```python
# Crear un imputador que utilice la moda (el valor más frecuente) para llenar los valores faltantes
imputer_mode = SimpleImputer(strategy='most_frequent')

# Lista de columnas categóricas que necesitan imputación de la moda
categorical_columns = ['education', 'BPMeds']

# Aplicar la imputación de la moda a las columnas especificadas
data[categorical_columns] = imputer_mode.fit_transform(data[categorical_columns])

# Verificar que no quedan valores faltantes en las columnas imputadas
print(data[categorical_columns].isnull().sum())
```

### Normalización
En este código:


*   Iteramos sobre las columnas continuas: Para cada columna, aplicamos el escalado y guardamos el resultado en una nueva columna en el mismo DataFrame, cuyo nombre será el original con el sufijo _scaled.
*   Descripción de las columnas escaladas: Al final, mostramos el resumen estadístico de las nuevas columnas escaladas para confirmar que la normalización se ha aplicado correctamente y que todos los valores están ahora entre 0 y 1.




```python
from sklearn.preprocessing import MinMaxScaler
# Crear el objeto MinMaxScaler
scaler = MinMaxScaler()

# Lista de columnas continuas que necesitan ser escaladas
continuous_columns = ['cigsPerDay', 'totChol', 'BMI', 'heartRate', 'glucose']

# Aplicar MinMaxScaler a las columnas especificadas y crear nuevas columnas para los resultados
for column in continuous_columns:
    data[column + '_scaled'] = scaler.fit_transform(data[[column]])

# Verificar que la normalización se ha aplicado correctamente a las nuevas columnas
print(data[[col + '_scaled' for col in continuous_columns]].describe())
```

## **4. Submuestras**
Realiza un análisis básico de selección de características siguiendo alguna de
las técnicas vistas a lo largo del módulo que te permita obtener una submuestra
depurada del dataset original.

### **Selección de Características usando Árboles de Decisión**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

# Asumiendo que 'TenYearCHD' es la variable objetivo
X = data.drop('TenYearCHD', axis=1)  # todas las características excepto la variable objetivo
y = data['TenYearCHD']  # la variable objetivo

# Crear el modelo RandomForest
forest = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)

# Entrenar el modelo
forest.fit(X, y)

# Crear un selector de características basado en la importancia calculada por el RandomForest
selector = SelectFromModel(forest, prefit=True)

# Reducir el conjunto de datos a las características más importantes
X_reduced = selector.transform(X)

# Convertir X_reduced de nuevo a DataFrame para mejor manejo y visualización
X_reduced_df = pd.DataFrame(X_reduced, columns=X.columns[selector.get_support()])

print("Características seleccionadas:", X_reduced_df.columns)
```

### **Análisis de Correlación**

```python
# Calcular la matriz de correlación
corr_matrix = data.corr()

# Visualizar la matriz de correlación
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f")
plt.title('Matriz de Correlación de Características')
plt.show()

# Considerar eliminar características con alta correlación entre sí
```

## **5. Entrenamiento**
Partiendo de esta submuestra final, genera los grupos de entrenamiento y
validación e imprime pos pantalla alguna características del mismo.

```python
from sklearn.model_selection import train_test_split

# Asumiendo que 'TenYearCHD' es la variable objetivo y que 'X_reduced_df' contiene las características seleccionadas
y = data['TenYearCHD']  # la variable objetivo

# Dividir los datos en submuestras de entrenamiento y validación
X_train, X_val, y_train, y_val = train_test_split(X_reduced_df, y, test_size=0.2, random_state=42, stratify=y)

# Imprimir características de las submuestras
print("Tamaño del conjunto de entrenamiento:", X_train.shape)
print("Tamaño del conjunto de validación:", X_val.shape)
print("\nDistribución de la variable objetivo en el conjunto de entrenamiento:")
print(y_train.value_counts(normalize=True))  # Mostrar la proporción de cada clase
print("\nDistribución de la variable objetivo en el conjunto de validación:")
print(y_val.value_counts(normalize=True))  # Mostrar la proporción de cada clase
```

**Importancia de la División:**
- Evitar el sobreajuste: Entrenar con un conjunto y validar con otro diferente ayuda a asegurar que el modelo no solo memorice los datos de entrenamiento, sino que también generalice bien a nuevos datos.
- Validación del Modelo: Permite evaluar la capacidad del modelo para funcionar efectivamente en datos no vistos durante el entrenamiento.

## **6. Sci-kit Learn**
Instancia el modelo Scikit-learn de tu elección y aplícalo sobre los datos
seleccionados para el entrenamiento.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Instanciar el modelo de regresión logística
logreg = LogisticRegression(random_state=42, max_iter=1000)

# Entrenar el modelo usando el conjunto de entrenamiento
logreg.fit(X_train, y_train)

# Realizar predicciones en el conjunto de validación
y_pred = logreg.predict(X_val)

# Evaluar el modelo
accuracy = accuracy_score(y_val, y_pred)
print("Precisión del modelo en el conjunto de validación:", accuracy)
print("\nReporte de clasificación:")
print(classification_report(y_val, y_pred))
```

Dado que estamos trabajando con un problema de clasificación para predecir el riesgo de enfermedad cardíaca en un período de diez años (TenYearCHD), un modelo simple y efectivo para comenzar sería la Regresión Logística. Este modelo es adecuado para problemas de clasificación binaria y proporciona resultados interpretables, lo cual es útil en contextos médicos donde entender la influencia de cada factor es crucial.

**Explicación del Código:**
- Instanciación del Modelo: LogisticRegression() es utilizado para instanciar el modelo. El parámetro max_iter=1000 se asegura de que el algoritmo tenga suficientes iteraciones para converger, lo cual puede ser necesario en conjuntos de datos más complejos o más grandes.
- Entrenamiento del Modelo: fit() se usa para entrenar el modelo con los datos de entrenamiento X_train y las etiquetas y_train.
- Predicción y Evaluación: Usamos predict() para hacer predicciones sobre el conjunto de validación X_val. Luego, evaluamos la precisión del modelo y mostramos un reporte de clasificación que incluye métricas como precisión, recall, y F1-score para cada clase.

**Importancia de la Evaluación:**
- Precisión: Nos da una idea de qué tan bien el modelo puede identificar correctamente a los pacientes en riesgo.
- Reporte de Clasificación: Proporciona una visión más detallada de la capacidad del modelo para manejar tanto la clase positiva (pacientes en riesgo) como la negativa (pacientes sin riesgo).



### Resultados:
- **Precisión General (Accuracy)**: 84.78%. Esto indica que, en general, el modelo predijo correctamente el 84.78% de los casos en el conjunto de validación.
- **Desempeño por Clase**:
  - **Clase 0 (No riesgo)**: Alta precisión (85%) y recall (99%), lo cual significa que el modelo es muy bueno prediciendo los casos de no riesgo.
  - **Clase 1 (Riesgo)**: Precisión del 50% y un recall muy bajo de 4%. Esto indica que aunque cuando el modelo predice que un paciente está en riesgo, tiene una probabilidad del 50% de estar en lo correcto, sólo detecta el 4% de todos los casos reales de riesgo.

### Interpretación y Problemas:
1. **Desequilibrio en el Rendimiento entre Clases**:
   - El modelo es eficiente identificando pacientes que no están en riesgo, pero falla significativamente en identificar correctamente a los pacientes que sí están en riesgo. Esto es problemático, especialmente en un contexto médico donde identificar a los pacientes en riesgo puede ser más crítico que identificar a los que no lo están.
   
2. **Posible Sesgo hacia la Clase Mayoritaria**:
   - La alta precisión y recall para la clase 0 sugiere que el modelo puede estar sesgado hacia la clase mayoritaria, lo cual es típico en conjuntos de datos desbalanceados. Esto puede llevar a una alta tasa de falsos negativos para la clase minoritaria (riesgo), como se ve reflejado en el bajo recall para la clase 1.

3. **F1-Score Bajo para la Clase 1**:
   - El F1-score combina tanto la precisión como el recall en una sola métrica. Un F1-score de 0.07 para la clase 1 es un indicador de que el modelo no está funcionando bien para esta clase.

### Recomendaciones para Mejoras:
- **Rebalanceo de las Clases**: Considera técnicas como el sobremuestreo de la clase minoritaria (SMOTE) o el submuestreo de la clase mayoritaria para equilibrar el conjunto de entrenamiento.
- **Ajustar el Umbral de Clasificación**: Experimentar con diferentes umbrales para la decisión de clasificación podría ayudar a mejorar el recall de la clase minoritaria sin sacrificar demasiado la precisión.
- **Evaluar Otras Métricas**: Dado el desequilibrio entre las clases, métricas como la curva ROC-AUC o el área bajo la curva de precisión-recall (PR AUC) podrían proporcionar una mejor comprensión del desempeño del modelo.
- **Probar Diferentes Modelos**: Algunos modelos pueden manejar mejor el desbalance de clases, como los modelos de ensamble (Random Forest, Gradient Boosting) o ajustar los parámetros de regularización y penalización en modelos lineales.

Estos pasos deberían ayudar a mejorar el modelo, haciéndolo más efectivo y equilibrado en la predicción de ambos tipos de resultados, especialmente para identificar correctamente a los pacientes en riesgo.

# **7. Conjunto de validación**
Calcula y muestra por pantalla información (métricas) sobre el desempeño del
modelo frente al conjunto de validación.  

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Precisión del modelo
accuracy = accuracy_score(y_val, y_pred)
print("Precisión del modelo en el conjunto de validación:", accuracy)

# Matriz de confusión
conf_matrix = confusion_matrix(y_val, y_pred)
print("\nMatriz de Confusión:")
print(conf_matrix)

# Reporte de clasificación
print("\nReporte de Clasificación:")
print(classification_report(y_val, y_pred))

# Curva ROC y AUC
fpr, tpr, thresholds = roc_curve(y_val, logreg.predict_proba(X_val)[:, 1])
auc = roc_auc_score(y_val, y_pred)
print("\nAUC (Área Bajo la Curva):", auc)

# Graficar la curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC')
plt.legend(loc="lower right")
plt.show()
```

**Descripción de las Métricas:**
- Precisión (Accuracy): Esta métrica mide la proporción de predicciones correctas (tanto positivas como negativas) en relación con todas las predicciones hechas.
- Matriz de Confusión: Muestra la cantidad de predicciones correctas e incorrectas divididas por clase. Los valores diagonales representan las predicciones correctas.
- Reporte de Clasificación: Incluye precisión (qué proporción de identificaciones positivas fue realmente correcta), recall (qué proporción de positivos reales fue identificada correctamente) y el F1-score (un promedio ponderado de la precisión y el recall).
- Curva ROC y AUC: La curva ROC muestra la relación entre la tasa de verdaderos positivos y la tasa de falsos positivos. El AUC proporciona un valor único que resume el rendimiento del modelo en todas las tasas de clasificación posibles; cuanto más cercano a 1, mejor es el modelo.
